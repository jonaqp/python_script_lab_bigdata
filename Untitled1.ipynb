```python
dftotal=spark.sql("""
                        select a.CustomerId
                                ,cast(a.TOB as decimal(8,2)) as TOB
                                ,cast(b.IncomeLevel as decimal(8,2)) as IncomeLevel
                                ,cast(a.Bal01 as decimal(8,2)) as Bal01
                                ,cast(a.MaxDqBin01 as decimal(8,2)) as MaxDqBin01
                                ,cast(a.MtgBal01 as decimal(8,2)) as MtgBal01
                                ,cast(a.NonBankTradesDq01 as decimal(8,2)) as NonBankTradesDq01
                                ,cast(a.FlagGB as int) as FlagGB
                                ,cast(a.FlagSample as int) as FlagSample
                         from mmedinac.data_in_chilean a
                    inner join mmedinac.data_in_clientes b on a.CustomerId = b.NUMERODOCUMENTO
                  """)
```

```python
dftotal.count()
```

```python
dftotal.createOrReplaceTempView("tabla_final")
dftrain=dftotal[dftotal['FlagSample']==1].select('CustomerId','TOB','MtgBal01')
```

```python
training_data=spark.sql("""
                    select  CustomerId, TOB, IncomeLevel, Bal01,a.MaxDqBin01,a.MtgBal01,a.NonBankTradesDq01,a.FlagGB
                      from tabla_final a
                    where FlagSample = 1
                  """)
test_data=spark.sql("""
                    select  CustomerId, TOB, IncomeLevel, Bal01,a.MaxDqBin01,a.MtgBal01,a.NonBankTradesDq01,a.FlagGB
                      from tabla_final a
                    where FlagSample = 0
                """)
```

```python
test_data.dtypes
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pandas.tools.plotting import scatter_matrix
%matplotlib inline 
import matplotlib.pyplot as plt, numpy as np
```

```python
assembler = VectorAssembler(inputCols=["TOB", "IncomeLevel", "Bal01","MaxDqBin01","MtgBal01","NonBankTradesDq01"], outputCol="features")
assem_data = assembler.transform(training_data)

assembler_test = VectorAssembler(inputCols=["TOB", "IncomeLevel", "Bal01","MaxDqBin01","MtgBal01","NonBankTradesDq01"], outputCol="features")
assem_data_test = assembler_test.transform(test_data)
```

```python

train_scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=True)
train_scaler_model = train_scaler.fit(assem_data)
scaled_data_train = train_scaler_model.transform(assem_data)

test_scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=True)
test_scaler_model = test_scaler.fit(assem_data_test)
scaled_data_test = test_scaler_model.transform(assem_data_test)

```

```python
rf = RandomForestClassifier(labelCol="FlagGB", featuresCol="scaled_features")
rfModel = rf.fit(scaled_data_train)
```

```python
#rfModel.write().overwrite().save("/user/epinedac/app/rfmodel")
```

```python
# Score test Data
predictions = rfModel.transform(scaled_data_test)
predictions.show()
```

```python
predictions.head(2)
```

```python
from pyspark.sql.functions import udf, col, round, expr
from pyspark.sql.types import FloatType

split1_udf = udf(lambda value: value[0].item(), FloatType())
split2_udf = udf(lambda value: value[1].item(), FloatType())

df_1_total_output = predictions.select("customerid","FlagGB","prediction",
                                                   split1_udf('probability').alias('p0'),
                                                   split2_udf('probability').alias('p1'))

df_1_total_output_exporta=df_1_total_output.selectExpr("*", "(case when prediction == 1 then p1 else p0 end) as probability")
```

```python
prediction_2=predictions.select(expr("*"),expr("cast(probability as string)  as probability_2"))
```

```python
prediction_final=prediction_2.select("customerid","FlagGB","prediction",expr("substr(probability_2,1,instr(probability_2,',')) as p0"),
                       expr("substr(probability_2,instr(probability_2,',')+1,length(probability_2)) as p1"),
                      )
```
